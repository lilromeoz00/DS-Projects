```python
import numpy as np
```

```python
# OR GATE PERCEPTRON
```

```python
input_features = np.array([[0,0],[0,1],[1,0],[1,1]])
print(input_features.shape)
input_features
```

```python
target_output = np.array([[0,1,1,1]])
target_output = target_output.reshape(4,1)
print(target_output.shape)
target_output
```

```python
weights = np.array([[.1],[.2]])
print(weights.shape)
weights
```

```python
bias = 0.3
lr = 0.05
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def sigmoid_der(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
for epoch in range(25000):
    inputs = input_features
    
    # Feedforward input
    in_o = np.dot(inputs, weights) + bias
    
    # Feedforward output
    out_o = sigmoid(in_o)
    
    # Backprop & Calc Error
    error = out_o - target_output
    
    x = error.sum()
    print(x)
    
    # Calc derivative
    derror_douto = error
    douto_dino = sigmoid_der(out_o)
    
    # Mult individual derivatives
    deriv = derror_douto * douto_dino
    
    # Find transpose of input features
    inputs = input_features.T
    deriv_final = np.dot(inputs,deriv)
    
    # Update Weights
    weights -= lr * deriv_final
    
    # Updating bias weight value
    for i in deriv:
        bias -= lr * i
    
```

```python
print(weights)
print(bias)
```

```python
# Inputs
single_point = np.array([1,0])

# Step 1
result1 = np.dot(single_point, weights) + bias

# Step 2
result2 = sigmoid(result1)

print(result2)

# Added code to round up
r2 = result2
if r2 >= .5:
    r2 = 1
else:
    r2 = 0
print(r2)
```
