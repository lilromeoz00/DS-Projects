```python
import tensorflow_hub as hub
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
# "Borrow" this pretrained model for a second
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
# Neat little function to process an img for us
def load_image(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    return img
```

```python
content_img = load_image(r'C:\Users\jordan\Desktop\jordan.png')
style_img = load_image(r'C:\Users\jordan\Desktop\noodles.jpg')
```

```python
content_img.shape
```

```python
plt.imshow(np.squeeze(style_img))
plt.show()
```

```python
stylized = model(tf.constant(content_img), tf.constant(style_img))[0]
```

```python
plt.imshow(np.squeeze(stylized))
plt.show()
```

```python
cv2.imwrite('jordanSpag.jpg', cv2.cvtColor(np.squeeze(stylized)*255, cv2.COLOR_BGR2RGB))
```
